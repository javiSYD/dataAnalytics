```python
#import libraries  
import numpy as np
import pandas as pd 
from transformers import AutoModelForSequenceClassification, BertForSequenceClassification
from transformers import pipeline
from transformers import TFAutoModelForSequenceClassification
from transformers import AutoTokenizer, AutoConfig
from scipy.special import softmax
```

```python
classifier = pipeline('sentiment-analysis')
```

```python
res = classifier('i got a job as a Data Analytics') 
```

```python
print(res)
```

```python
df = pd.read_csv('cleanDf.csv')
```

```python
df.head()
```

```python
#extract the column Messages to have it as an input for the sentiment analysis 
messages = df['Messages']
```

```python
#from the 1.5 million conversations we keep only 493,396 for the sentiment ansalysis. 
sentiment = []
sentScore = []
for message in messages:
    output = classifier(message)
    label = output[0]['label']
    score = output[0]['score']
    sentiment.append(label)
    sentScore.append(score)
```

```python
#create a copy of the data frame
newDf = df.copy()
```

```python
#keep 493.396 rows to run our linear model
newDf.drop(index=newDf.index[493395:-1],axis=0,inplace=True)
```

```python
#
newDf.insert(3, 'Sentiment', sentiment)
newDf.insert(4, 'SentScore', sentScore)
```

```python
#export to csv
newDf.to_csv('sentimentAnalysis .csv', encoding='utf-8')
```
