```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#create numbers from 0 to 5 
x = np.linspace(0,5,11)
y = x **2
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r*')
plt.show()
```

```python
plt.hist(y)
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python
 plt.scatter(y,x)
 plt.show()
```

```python
plt.boxplot(y)
plt.show()
```

SUBPLOT 

```python
# More with than graph 
#(number of rows, number of columns, index between rows and columns)
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

```python
x = np.linspace(0,5,11)
y = x **2
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y,'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes2 = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y,'b')
axes2.plot(y,x,'r')
```
