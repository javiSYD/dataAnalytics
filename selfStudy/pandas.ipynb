```python
def two_times(value):
    return value * 2
```

```python
#For each element in the colum 'User Rating' we will apply the function 
df_books['Rating 2'] = df_books['User Rating'].apply(two_times)
df_books
```

```python
#I can do the same with a lamda function 
df_books['Rating 2'] = df_books['User Rating'].apply(lambda x : x * 3)
df_books.head()
```

```python
#Use diferent colums to make the operations 
# None Fiction is the same but fiction is multiply by 2 
df_books.apply(lambda x : x['User Rating']*2 if x['Genre'] == 'Fiction' else x ['User Rating'],axis = 1)
```
