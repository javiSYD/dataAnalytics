```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig = plt.figure()
                 #(position in the figure, size of the axes)
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y,'b')
```

```python
#Change each object inside the figure 
fig = plt.figure()
                 #(position in the figure, size of the axes)
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.6,0.4,0.3])

axes.plot(x,y,'b')
axes2.plot(y,x,'r')

axes2.set_facecolor('gray')
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

SUBPLOTS

```python
fig, axes = plt.subplots()
axes.plot(x,y, 'b')
```

```python
                      #(one column with 2 graphs)
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(y,x, 'b')
axes[1].plot(x,y, 'r')
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(y,x, 'b')
ax2.plot(x,y, 'r')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=4)
axes[0,0].plot(np.tan(x),y,'r')
axes[0,1].plot(y,x, 'b')
axes[0,2].plot(np.sin(y),x,'g')
axes[0,3].plot(x,y, 'r')
axes[1,0].plot(np.sin(x),y,'y')
axes[1,2].plot(y,x,'g')


# make each graph with a better distribution 
fig.tight_layout()
```

```python
rows = 3
cols = 3
fig, axes =plt.subplots(nrows=rows, ncols=cols, figsize=(12,6))
for row in range(rows):
    for col in range(cols):
        axes[row,col].plot(x,np.sin(x))
        axes[row,col].set_title('sin(x)')
fig.tight_layout()
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

LABLES

```python
fig, axes = plt.subplots(1,2, figsize=(8,8))
axes[0].plot(x,y, label='$sin(x)$')
axes[0].set_title('Relacion X-Y')
axes[0].set_xlabel('X')
axes[0].legend(loc='lower right')

axes[1].plot(y,x)
axes[1].set_title('Relacion Y-X')
axes[1].set_ylabel('Y')
```
